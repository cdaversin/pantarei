```python
dt = 0.5
T = 10
time = TimeKeeper(dt, T)
```

```python
domain = MMSDomain(13)
p_expr = df.Expression("x[0]*x[0]", degree=2, domain=domain)
u_expr = df.Expression("(1 - t / T) * x[0]*x[0]", t=time, T=T, degree=2, domain=domain)
V = df.FunctionSpace(domain, "CG", 1)
p = df.project(p_expr, V)
p.rename("pressure", "")

u = df.project(u_expr, V)
u.rename("concentration", "") 

store = FenicsStorage(Path("../mytestfile.hdf5"), "w") 
store.write_domain(domain)
store.write_function(p)
store.write_function(u)
for idx, ti in enumerate(time):
    u.assign(df.project(u_expr, V))
    store.write_checkpoint(u, idx+1)
store.write_timevector(time.as_vector(), u.name())
store.close()

reader = FenicsStorage(Path("../mytestfile.hdf5"), "r")
P = reader.read_function("pressure")
df.plot(P)
plt.title("Pressure")
plt.show()

t = reader.read_timevector("concentration")
U = reader.read_function("concentration")
for idx, ti in enumerate(t):
    U = reader.read_checkpoint(u, idx)
    df.plot(U, vmin=0, vmax=2)
    plt.title(f"Concentration t={ti}")
    plt.show()

reader.close()
```

# MixedElements

```python
time.reset()
cell_type = 'triangle'
QV = df.VectorElement("CG", ufl.Cell(cell_type), 2)
LP = df.FiniteElement("CG", ufl.Cell(cell_type), 1)
mixed_element = df.MixedElement([QV, LP])
W = df.FunctionSpace(domain, mixed_element)
mixed_func = df.Function(W)
data = {
    0: df.Function(W.sub(0).collapse()),
    1: df.project(p_expr, W.sub(1).collapse()),
}
mixed_func = assign_mixed_function(data, W, data)
```

```python
def assign_mixed_function(p, V, compartments):
    """Create a function in a mixed function-space with sub-function being
    assigned from a dictionray of functions living in the subspaces."""
    P = df.Function(V)
    subspaces = [V.sub(idx).collapse() for idx, _ in enumerate(compartments)]
    Pint = [df.interpolate(p[j], Vj) for j, Vj in zip(compartments, subspaces)]
    assigner = df.FunctionAssigner(V, subspaces)
    assigner.assign(P, Pint)
    return P
```
